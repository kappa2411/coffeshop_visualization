```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_excel(r'C:\sqldata\coffee_house_dataset_500.xlsx')
df['Date'] = pd.to_datetime(df['Date'])
daily_sales = df.groupby('Date')['Total_Bill'].sum()

plt.plot(daily_sales)
plt.title('Total Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()

```

```python


category_sales = df.groupby('Coffee_Type')['Total_Bill'].sum()

category_sales.plot(kind='bar')
plt.title('Sales by Category')
plt.show()
```

```python

top_items = df.groupby('Coffee_Type')['Quantity'].sum().sort_values(ascending=False).head(5)

top_items.plot(kind='barh')
plt.title('Top 5 Best Selling Items')
plt.show()
```

```python

category_sales.plot(kind='pie', autopct='%1.1f%%')
plt.title('Category Contribution')
plt.show()
```

```python

plt.hist(df['Total_Bill'], bins=10)

plt.title('Distribution of Order Values')
plt.xlabel('Order Value')
plt.ylabel('Frequency')

plt.show()
```

```python

import seaborn as sns
sns.lineplot(x=daily_sales.index, y=daily_sales.values)
plt.show()
```

```python

sns.barplot(x=category_sales.index, y=category_sales.values)
plt.show()
```

```python

sns.histplot(df['Total_Bill'], bins=10)

plt.title('Distribution of Order Values / Total Sales')
plt.xlabel('Order Value (Total Price)')
plt.ylabel('Frequency')

plt.show()
```

```python

sns.set_style("whitegrid")
category_sales = df.groupby('Coffee_Type')['Total_Bill'].sum()
plt.figure(figsize=(6,6))
plt.pie(category_sales, labels=category_sales.index, autopct='%1.1f%%', startangle=140)
plt.title('Percentage Contribution of Each Category')
plt.show()
```
